## Finite Differences for the Heat Equation

Consider the initial-boundary value problem for the heat equation:

$$ u_t = Du_{xx} \qquad 0 < x < \pi $$
$$ D > 0, \qquad u(t, 0) = u(t,\pi) = 0$$

and 

\begin{equation}
u(0, x)=
    \begin{cases}
        x & \text{for } 0 \leq x \leq \pi/2 \\
        \pi - x  & \text{for } \pi/2 \leq x \leq \pi 
    \end{cases}
\end{equation}


We are given the explicit finite difference scheme

$$ \frac{u_{j}^{n+1} - u_{j}^{n} }{\Delta t} = D \frac{ u_{j-1}^{n} - 2u_{j-1}^{n} + u_{j-1}^{n} }{(\Delta x)^2} $$

for $j = 1,2,3, \ldots M-1$ and $n=0,1,2,3, \ldots N-1$, and $u_{0}^{N} = u_{M}^{n} = 0$

We will write a code to implement this finite difference method and, by taking suitable values of $\alpha = D\Delta t/(\Delta x)^2$, and test the necessary condition for stability. We will also consider one value of $\alpha$ for which the scheme is stable and one for which it is not and lot the corresponding approximate solutions.

To implement this finite difference method, first we need to write the scheme as follows

$$ u_{j}^{n+1} = u_{j}^{n} + \alpha \big[u_{j+1}^{n} - 2u_{j}^{n} + u_{j-1}^{n} \big] $$

Now we write the code using the one step method

```python
# Importing the libraries
import numpy as np
import math
from copy import deepcopy
import matplotlib.pyplot as plt
```

```python
# Initializing the IVP
M = 20
D = 1
alpha = 0.45
N = 1000

delta_x = math.pi/ M
delta_t = alpha * (delta_x ** 2)/D
u_0 = np.zeros(M+1)

# Initializing u_0
def calculate_u_init(M, delta_x):
    for i in range(0, M+1):
        x = i * delta_x
        if x < math.pi/2:
            u_0[i] = x
        else:
            u_0[i] = math.pi - x
    return u_0

u_init = calculate_u_init(M, delta_x)
print(u_init)

x_values = np.zeros(M+1)
for i in range(0, M+1):
    x_values[i] = i * delta_x 
    


# Function for calculating explicit forward difference 
def explicit_fd(u_init, alpha, N):
    current_approx = u_init
    
    for _ in range(1, N+1):
        next_approx = np.zeros_like(current_approx)
    
        for j in range(1, M):
            next_approx[j] = current_approx[j] + \
                alpha *(current_approx[j-1] - 2 * current_approx[j] + current_approx[j+1])
            
        current_approx = next_approx
        
    return current_approx

u_approx = explicit_fd(u_init, alpha, N)

# Plotting the graphs to check stability when alpha < 0.5
plt.plot(x_values, u_init)
plt.plot(x_values, u_approx)
```

```python
alpha = 0.55
u_init = calculate_u_init(M, delta_x)
u_approx = explicit_fd(u_init, alpha, N)

# Plotting to check stability when alpha > 0.5
plt.plot(x_values, u_init)
plt.plot(x_values, u_approx)
```

We can confirm that the plot is unstable when alphs is greater than 0.5 and stable when it is less than 0.5. We can change the values of $\alpha$ and we will see that $0.5$ is the tipping point for the system to be stable.
