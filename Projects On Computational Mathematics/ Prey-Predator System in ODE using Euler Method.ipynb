# Prey-Predator System in ODE

Consider the ODE system

$$ x' =x(1-y) $$
$$ y' =y(x-1)$$

This is a simple model in for a prey-predator system. Here $x(t)$ and $y(t)$ measure (in dimensionless form) the size of the prey and the predator populations, respectively. Implement the improved Euler method below,

\begin{equation}
\begin{split}
K_1 &= f(t_, y^n)\\
K_2 &= f(t_{n+1}, y^n + \Delta t K_1)\\
y^{n+1} &= y^n + \frac{\Delta t}{2}[k_1 + K_2]
\end{split}
\end{equation}

to find an approximation to the solution of this system for $0 \leq t \leq T$. Take $T = 8$ and use $\Delta t=0.01$ to compute the approximation for the following initial conditions (a) $x(0) = 0.25$, $y(0) = 0.75$, (b) $x(0) = 0.5$, $y(0) = 0.5$, (c) $x(0) = 0.90$, $y(0) = 0.90$ and plot the curves $(x(t),y(t))$. What do these curves tell you about the behavior of this prey-predator system?

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Step size
h = 0.01
t = np.arange(0, 8, h)
n = len(t)
# print(n)
x = np.zeros(n)
y = np.zeros(n)

# Initial values of x_0 and y_0
x[0] = 0.25
y[0] = 0.75
```

```python
def ODE_system(x, y):
    diff_x = x * (1-y)
    diff_y = y * (x-1)
    return diff_x, diff_y
```

```python
K_1_for_x = np.zeros(n)
K_1_for_y = np.zeros(n)
K_2_for_x = np.zeros(n)
K_2_for_y = np.zeros(n)

def improved_euler_method(x, y, t):
    for i in range(0, n-1):
        K_1_for_x[i], K_1_for_y[i] = ODE_system(x[i], y[i])
        K_2_for_x[i], K_2_for_y[i] = ODE_system(x[i] + h * K_1_for_x[i], y[i] + h * K_1_for_y[i] )
        x[i+1] = x[i] + (h/2) * ( K_1_for_x[i] + K_2_for_x[i] )
        y[i+1] = y[i] + (h/2) * ( K_1_for_y[i] + K_2_for_y[i] )
```

```python
improved_euler_method(x,y,t)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(x, y, label='Prey-Predator Population')
plt.title('Solutions Curves to the ODE')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```

```python
x[0] = 0.5
y[0] = 0.5

K_1_for_x = np.zeros(n)
K_1_for_y = np.zeros(n)
K_2_for_x = np.zeros(n)
K_2_for_y = np.zeros(n)
improved_euler_method(x,y,t)

plt.figure(figsize = (12, 8))
plt.plot(x, y, label='Prey-Predator Population')
plt.plot(0.5, 0.5, marker = "o")
plt.title('Solutions Curves to the ODE')
plt.xlabel('Time')
plt.ylabel('Population')
plt.grid()
plt.legend()
plt.show()
```

```python
x[0] = 0.9
y[0] = 0.9

K_1_for_x = np.zeros(n)
K_1_for_y = np.zeros(n)
K_2_for_x = np.zeros(n)
K_2_for_y = np.zeros(n)
improved_euler_method(x,y,t)

plt.figure(figsize = (12, 8))
plt.plot(x, y, label='Prey-Predator Population')
plt.plot(0.9, 0.9, marker = "o")
plt.title('Solutions Curves to the ODE')
plt.xlabel('Time')
plt.ylabel('Population')
plt.grid()
plt.legend()
plt.show()
```

Looking at the curves, we can infer the behavior of prey-predator system is periodical. After a certain time, the prey-predator populations tend to return to their intial values.
