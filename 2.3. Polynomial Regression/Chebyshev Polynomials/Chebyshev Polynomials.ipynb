# Chebyshev Polynomials

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def chebyshev_polynomials(n):
    x = np.linspace(-1, 1, 100)
    plt.xlim([-1.2, 1.2])
    #plt.ylim([-1.2, 1.2])
    chebyshev_poly = np.cos(n * np.arccos(x))
    plt.plot(x, chebyshev_poly)
    plt.show()



T_1 = chebyshev_polynomials(1)
T_2 = chebyshev_polynomials(2)
T_3 = chebyshev_polynomials(3)
T_4 = chebyshev_polynomials(4)
T_5 = chebyshev_polynomials(5)
```
