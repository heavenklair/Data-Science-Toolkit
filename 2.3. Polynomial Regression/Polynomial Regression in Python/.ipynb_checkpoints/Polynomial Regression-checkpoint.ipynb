# Polynomial Linear Regression

### Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

We do not want to take into account the first column of the dataset because it is redundant. The second column already provides a categorical values for the first one, so dropping the first column does not effect anything in the preprocessing stage.

### Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

### Training the Polynomial Linear Regression on the whole dataset 

The first step is to import the tool that will allow us to create a matrix of $x^i$, where $0 \leq i \leq n$. $n$ is the degree we declare in the PolynomialFeature below. 

In "poly_reg.fit_transform(X)", X is the array of levels in the dataset. We want to take this array and transform into a matrix of different powers of X. That is what this transform method is doing here, returing a matrix of features.

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

Now we have a new matrix of features, composed of $x_1$ and $x_1^2$. We have to build a linear regression model that will integrate these features into the equation
$$ y = b_0 + b_1x_1 + b_2x_1^2 + \ldots + b_nx_1^n $$
This is where the "Linear" comes in in "Polynomial Linear Regression". Thus, the lines 4 and 5.

### Visualising the Linear Regression Results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regresion)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

### Visualising the Polynomial Regression Results

How do we need to change in the Polynomial Linear Regression?

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Linear Regresion)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

We arbitrarily choose $n = 2$. What if we increase the degree of polynomial function. Intuition says that we will get a better and higher resolution of the graph

### Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Linear Regresion)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

The above code is for appreciation of the smoother curves

### Predicting a new result with Linear Regression

We will predicting the result for value of X = 6.5

```python
lin_reg.predict([[6.5]])
```

### Predicting a new result with Polynomial Regresssion

What do we need to change in the code for our prediction using the Polynomial Linear Regression?

We cannot input a single value like the Linear Regression. So, we need to input the features of the equation i.e. $x_1^i$, different values of power of $x_1$.

For that, we can use the "poly_reg.fit_transform( [ [6.5] ] )" to get a increasing powers of $x$ 

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```

### Case Study
Let say the dataset used in this modeling is of a person who has had held different positions a company. Right now, he is at level 6. He wanted to change his company for better pay, so he went to interview with this new company. The interviewer asked him what is his salary expectation? He said "At least 160K". Interviewer asked "why this number?". He replied "Because I was getting paid 160K at my previous company" 

How can we make sure that he is saying the correct number? 

We can perform our linear regression and Polynomial Linear Regression to see that Polynomial Linear Regression gives us the better and close result which confirms that yes, he is saying the truth.
